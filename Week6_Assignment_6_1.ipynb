```python
from sklearn.datasets import load_iris
import pandas as pd


# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
#Print the first 5 rows to observe your data
iris_df.head()
```

1. Compute mean, median, and mode for each feature.

```python
for column in iris_df.columns:
    mean = iris_df[column].mean()
    median = iris_df[column].median()
    mode = iris_df[column].mode()[0]
    print(f"Feature: {column}")
    print(f"  Mean: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")
```

2. Calculate the range, variance, and standard deviation for each feature.

```python
range_df = iris_df.max() - iris_df.min()
variance_df = iris_df.var()
std_dev_df = iris_df.std()

print("Range for each feature:\n", range_df)
print("\nVariance for each feature:\n", variance_df)
print("\nStandard Deviation for each feature:\n", std_dev_df)
```

3. Compute the 25th, 50th, and 75th percentiles (quartiles) for each feature.

```python
percentiles = iris_df.quantile([0.25, 0.5, 0.75])

print("25th, 50th, and 75th percentiles for each feature:\n", percentiles)
```

4. Create a histogram for each of the features to visualize their distributions.

```python
import matplotlib.pyplot as plt

for column in iris_df.columns:
    plt.figure()
    plt.hist(iris_df[column], bins=20, edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

5. Plot scatter plots between pairs of features to explore relationships.

```python
pd.plotting.scatter_matrix(iris_df, figsize=(12, 12), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)

plt.show()
```
