Take an input of numbers in a list and create a new list containing only the even numbers from the original list.

```python
#Your answer must go here

User_Input = input("Enter numbers separated by spaces: ").split()


Numbers = [int(num) for num in User_Input]

Even_list = [num for num in Numbers if num % 2 == 0]


print(Even_list)
```
